```python
import numpy as np 
import matplotlib.pyplot as plt 
%matplotlib inline 
```

### Computation of Fourier Transform 
#### Example 1 - single real sinusoids 

```python
# sampling freq. (number of samples per sec) 
Fs = 1000 

# sampling period 
Ts = 1/Fs 

# freq. in Hz (number of cycles per sec)
f0 = 20  

# timestamp when sampling occurs for data points generation 
t = np.arange(0, 1, Ts) 

# total number of samples
tot_samples = len(t)  

# create a signal
cos_t = np.cos(2*np.pi*f0*t)
```

```python
plt.figure(figsize=(10, 20))
plt.subplot(411)
plt.plot(t, cos_t)
```

```python
# compute FT 
X_f = np.fft.fft(cos_t) 

# compute magnitude of FT 
ft_mag = abs(X_f) 

f = np.arange(0, Fs, Fs/tot_samples) 
plt.subplot(412)
plt.plot(f, ft_mag)

# the frequency axis including negative frequencies
f2 = np.arange(-Fs/2, Fs/2, Fs/tot_samples)  
plt.subplot(413)
ft_shifted = np.fft.fftshift(ft_mag)
plt.plot(f2, ft_shifted) 

# the frequency axis excluding negative frequencies
f3 = np.arange(0, Fs/2, Fs/tot_samples)  
plt.subplot(414)
plt.plot(f3, ft_shifted[(int(Fs/2)):])
```

#### Example 2 - addition of two real sinusoids

```python
# define required parameters 

Fs = 100 # number of samples per sec 
Ts = 1 / Fs # sampling period 

# frequencies of the signals 
f0_1 = 10 
f0_2 = 20 

# timestamp when sampling occurs for generating data points 
t = np.arange(0, 1, Ts) 

# number of samples 
tot_samples = len(t)

# create the signals 
cos_t_0 = np.cos(2*np.pi*f0_1*t)
cos_t_1 = np.cos(2*np.pi*f0_2*t)

# create another signal by adding them
x_t = cos_t_0 + cos_t_1 
```

```python
# plot the signals 
plt.figure(figsize=(10, 10))
plt.subplot(311)
plt.plot(t, cos_t_0)

plt.subplot(312)
plt.plot(t, cos_t_1)

plt.subplot(313)
plt.plot(t, x_t)
```

```python
# compute the FT of the created signal 
X_f = np.fft.fft(x_t) 

# Compute spectrum / magnitude
ft_mag = abs(X_f) 

f = np.arange(0, Fs, Fs/tot_samples) 
plt.figure(figsize=(10, 10))

plt.subplot(311)
plt.plot(f, ft_mag)

# the frequency axis including negative frequencies 
f2 = np.arange(-Fs/2, Fs/2, Fs/tot_samples)  
plt.subplot(312)
ft_shifted = np.fft.fftshift(ft_mag)
plt.plot(f2, ft_shifted) 

# the frequency axis excluding negative frequencies
f3 = np.arange(0, Fs/2, Fs/tot_samples)  
plt.subplot(313)
plt.plot(f3, ft_shifted[(int(Fs/2)):])
```
